```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import preprocessing

import warnings
warnings.filterwarnings('ignore')
```

# Q1 Importing the data

```python
df = pd.read_csv('we_fn_use_c_marketing_customer_value_analysis.csv')
```

```python
df_num = df.select_dtypes(np.number)
```

```python
df_num.head()
```

```python
sns.distplot(df['Income'])
sns.distplot(df['Number of Policies'])
```

# Q2 Plotting the numerical variables

```python
sns.set(rc={'figure.figsize':(20,25)})
for i, column in enumerate(df_num.columns, 1):
    plt.subplot(4,2,i)
    sns.distplot(df_num[column])
```
